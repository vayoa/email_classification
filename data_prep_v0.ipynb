```python
# https://www.kaggle.com/datasets/jackksoncsie/spam-email-dataset
df = pd.read_csv('./spam_emails.csv')
```

```python
# https://huggingface.co/datasets/jason23322/high-accuracy-email-classifier

ndf = pd.read_csv("./full_dataset.csv").drop(columns=["id", "text"])
ndf["text"] = "Subject: " + ndf["subject"] + "\n\n" + ndf["body"]
ndf["important"] = ndf["text"].str.contains(
    "important|urgent|urgency|payment|security|notice|fail|expire|expired|password|invoice",
    case=False,
)
ndf.loc[ndf["important"] & (ndf["category"] == "spam"), "important"] = False
ndf
```

```python
print('Size:', len(ndf))
pd.concat(
    [
        (ndf["category"].value_counts(normalize=True) * 100).round(2).rename('percentage'),
        ndf["category"].value_counts().rename('count'),
    ],
    axis=1,
)
```
